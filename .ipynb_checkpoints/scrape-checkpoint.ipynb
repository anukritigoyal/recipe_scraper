```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame
```

```python
# using beautiful soup to scrape websites
r = requests.get("https://www.williams-sonoma.com/recipe/main-courses/beef-and-veal/?page=viewall")
soup = BeautifulSoup(r.text, 'html.parser')
scrape = soup.find_all('div', attrs={'class':'recipe-list'}) 
scrape_recipes = scrape[0].find_all('li')
```

```python
url = []
name = []
ingredient = []
```

```python
for recipes in scrape_recipes:
    recipe_info = recipes.contents
    link_to_recipe = recipe_info[1]['href']
    recipe_name = recipes.get_text()
    
    ingredient_request = requests.get(link_to_recipe)
    soup2 = BeautifulSoup(ingredient_request.text, 'html.parser')
    scrape2 = soup2.find_all('div', attrs={'class':'botLeft'})
            
    ingredients = scrape2[0].find_all('li')
    for ingred in ingredients:
        url.append(link_to_recipe)
        name.append(recipe_name)
        ingredient.append(ingred.get_text())
```

```python
df = pd.DataFrame({'url' : url, 'name' : name, 'ingredient' : ingredient})
# names actually do appear in csv, remove whitespace
df.to_csv('scraped_data.csv', index=False)
```
