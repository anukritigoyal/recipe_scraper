```python
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('cleaned_data.csv')
```

```python
# separating each word and making it its own entity
words = []
test = data['ingredient'][0]
for food in data['ingredient']:
    ingred = str(food)
    ingred = ingred.split(' ')
    for word in ingred:
        if word != "":
            words.append(word)
```

```python
counts = Counter(words)
top_ten = counts.most_common(10)
```

```python
calculate_data = pd.DataFrame(top_ten, columns=('word', 'count'))
```

```python
# calculating proportion
total = calculate_data['count'].sum()
proportion = []
for count in calculate_data['count']:
    value = count / total
    proportion.append(value)
calculate_data['proportion'] = proportion
```

```python
calculate_data.to_csv('calculations.csv', index=False)
```
