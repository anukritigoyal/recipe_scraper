```python
import pandas as pd
import re
```

```python
data = pd.read_csv('scraped_data.csv')
```

```python
# cleaned names
names = []
for name in data['name']:
    name = name.replace('\n', '')
    name = name.split('(')[0]
    
    names.append(name)
data['name'] = names
```

```python
# clean ingredients
ingredient = []
for ingred in data['ingredient']:
    ingreds = ingred.split(',')
    ingreds = ingreds[0].split('>')
    parens_index = ingreds[0].find(')')
    if parens_index != -1:
        ingreds = ingreds[0][parens_index + 2:]
        ingredient.append(ingreds)
    else:
        ingredient.append(ingreds[0])
```

```python
# removes measurement things ending with .
# removes measurements numbers in beginning
# removes any non-required words

ingredients = []
for ingred in ingredient:
    period_index = ingred.find('.')
    if period_index != -1:
        ingred = ingred[period_index + 2:]
    ingred = ingred.lower()
    ingred = ingred.replace('cups', '')
    ingred = ingred.replace('cup', '')
    ingred = ingred.replace('\xa0', '')
    ingred = ingred.replace('for serving', '')
    ingred = ingred.replace('for brushing', '')
    ingred = ingred.replace('for garnish', '')
    ingred = ingred.replace('for searing', '')
    ingred = ingred.replace('juice of 1', '')
    ingred = ingred.replace('grated', '')
    ingred = ingred.replace('zest', '')
    ingred = ingred.replace('peeled', '')
    ingred = ingred.replace('finely', '')
    ingred = ingred.replace('freshly', '')
    ingred = ingred.replace('fresh', '')
    ingred = ingred.replace('julienned', '')
    ingred = ingred.replace('small', '')
    ingred = ingred.replace('gluten-free', '')
    ingred = ingred.replace('sheet of', '')
    ingred = ingred.replace('dried', '')
    ingred = ingred.replace('thinly', '')
    ingred = ingred.replace('thin', '')
    ingred = ingred.replace('each', '')
    ingred = ingred.replace('large', '')
    ingred = ingred.replace('of 1 ', '')
    ingred = ingred.replace('of 1/2', '')
    ingred = ingred.replace('sliced', '')
    ingred = ingred.replace('dissolved in 3⁄4  boiling water', '')
    ingred = ingred.replace('plus 1 tsp.', '')
    ingred = ingred.replace('firmly packed', '')
    ingred = ingred.replace('1-inch-thick', '')
    ingred = ingred.replace('slices', '')
    ingred = ingred.replace('slice', '')
    ingred = ingred.replace('low-sodium', '')
    ingred = ingred.replace('pinch of ', '')
    ingred = ingred.replace('with masa harina in an electric pressure cooker', '')
    ingred = ingred.replace('with masa harina in a stovetop pressure cooker', '')
    ingred = ingred.replace('for dusting', '')
    ingred = ingred.replace('roughly chopped', '')
    ingred = ingred.replace('ready-to-eat', '')
    ingred = ingred.replace('shelled', '')
    ingred = ingred.replace('chipotle chilies in', '')
    ingred = ingred.replace('asian', '')
    ingred = ingred.replace('tri-tip', '')
    ingred = ingred.replace('suggested condiments: ', '')
    ingred = ingred.replace('fat-free', '')
    ingred = ingred.replace('shredded', '')
    ingred = ingred.replace('soft', '')
    ingred = ingred.replace('slivered', '')
    ingred = ingred.replace('grilled', '')
    ingred = ingred.replace('&', '')
    ingred = ingred.replace('(2 bay leaves', '')
    ingred = ingred.replace('roasted', '')
    ingred = ingred.replace('ground', '')
    ingred = ingred.replace('with air-fried ', '')
    ingred = ingred.replace('coarsely', '')
    ingred = ingred.replace('thick-cut', '')
    ingred = ingred.replace('thick', '')
    ingred = ingred.replace('chopped', '')
    ingred = ingred.replace('flat-leaf', '')
    ingred = ingred.replace('light', '')
    ingred = ingred.replace('bunch', '')
    ingred = ingred.replace('piece', '')
    ingred = ingred.replace('tbs.', '')
    ingred = ingred.replace('tb', '')
    ingred = ingred.replace('a 9- to 10-inch fireproof', '')
    ingred = ingred.replace('to ', '')
    ingred = ingred.replace('about 3 1/2inches deep', '')
    ingred = ingred.replace('herb bouquet:', '')
    ingred = ingred.replace('plus', '')
    ingred = ingred.replace('-inch', '')
    ingred = ingred.replace('round', '')
    ingred = ingred.replace('one ', '')
    ingred = ingred.replace('and ', '')
    ingred = ingred.replace('(about1 1/3 )', '')
    ingred = ingred.replace('kosher', '')
    ingred = re.sub(r'^[^a-zA-Z]*', '', ingred)
    ingredients.append(ingred)
```

```python
# removing anything before or and the word or itself
removed_or = []
for ingred in ingredients:
    or_index = ingred.find('or ')
    if or_index != -1:
        ingred = ingred[or_index + 3:]
    ingred = re.sub(' +',' ', ingred)
    ingred = re.sub('\d+', '', ingred)
    
    removed_or.append(ingred)
```

```python
data['ingredient'] = removed_or
```

```python
data.dropna(axis=0)
```

```python
data.to_csv('cleaned_data.csv', index=False)
```
